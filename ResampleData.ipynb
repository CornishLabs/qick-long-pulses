```python
# Resample data from Guido plot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data from PlotDigitiser
df_in = pd.read_csv("plot-data.csv")
df_in = df_in.sort_values('x')
```

```python
def resample_df(df_in: pd.DataFrame, new_x: np.ndarray) -> pd.DataFrame:
    # assume first col is x, second col is y
    x_name, y_name = df_in.columns[:2]

    x = df_in.iloc[:, 0].to_numpy()
    y = df_in.iloc[:, 1].to_numpy()

    # sort by x (keeps it simple and safe for np.interp)
    idx = np.argsort(x)
    x_sorted = x[idx]
    y_sorted = y[idx]

    new_y = np.interp(new_x, x_sorted, y_sorted)  # linear interp

    return pd.DataFrame({x_name: new_x, y_name: new_y})

df_out = resample_df(df_in, np.linspace(0,1,40))
```

```python
arr= df_out['y'].values
arr[arr<0.06]=0
df_out['y']=arr
np.in
```

```python
fig,ax = plt.subplots()
ax.plot(df_in['x'].values,df_in['y'].values)
ax.plot(df_out['x'],df_out['y'].values)
```
